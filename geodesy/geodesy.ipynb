```python
# 大地線正算
a = 6378245
f = 1/298.3
b = a*(1-f)
e2 = 1-b**2/a**2
e_prime2 = (a**2-b**2)/b**2
B_1 = 30
B_1 = B_1*pi/180
L_1 = 107
L_1 = L_1*pi/180
A_1 = 225
A_1 = A_1*pi/180
S = 1000000
W_1 = sqrt(1-e2*(sin(B_1)**2))
sinu_1 = sin(B_1)*sqrt(1-e2)/W_1
cosu_1 = cos(B_1)/W_1
sinA_0 = cosu_1*sin(A_1)
cosA_0 = sqrt(1-sinA_0**2)
cotsigma_1 = cosu_1*cos(A_1)/sinu_1
sin2sigma_1 = 2*cotsigma_1/(cotsigma_1**2+1)
cos2sigma_1 = (cotsigma_1**2-1)/(cotsigma_1**2+1)
k = sqrt(e_prime2*(cosA_0**2))
A = b*(1+k**2/4-3*(k**4)/64+5*(k**6)/256)
B = b*(k**2/8-k**4/32+15*(k**6)/1024)
C = b*(k**4/128-3*(k**6)/512)
alpha = (e2/2+e2**2/8+e2**3/16)-(e2**2/16+e2**3/16)*(cosA_0**2)+3*(e2**3)*(cosA_0**4)/128
beta = (e2**2/32+e2**3/32)*(cosA_0**2)-(e2**3/64)*(cosA_0**4)
sigma_0 = (S-(B+C*cos2sigma_1)*sin2sigma_1)*1/A
sin2 = sin2sigma_1*cos(2*sigma_0)+cos2sigma_1*sin(2*sigma_0)
cos2 = cos2sigma_1*cos(2*sigma_0)-sin2sigma_1*sin(2*sigma_0)
sigma = sigma_0+(B+5*C*cos2)*sin2/A
delta = (alpha*sigma+beta*(sin2-sin2sigma_1))*sinA_0
sinu_2 = sinu_1*cos(sigma)+cosu_1*cos(A_1)*sin(sigma)
B_2 = atan(sinu_2/(sqrt(1-e2)*sqrt(1-sinu_2**2)))
lamda = atan(sin(A_1)*sin(sigma)/(cosu_1*cos(sigma)-sinu_1*sin(sigma)*cos(A_1)))
if sin(A_1)>=0 and tan(lamda)>=0:
  lamda = abs(lamda)
elif sin(A_1)>=0 and tan(lamda)<0:
  lamda = pi-abs(lamda)
elif sin(A_1)<0 and tan(lamda)<0:
  lamda = -abs(lamda)
else:
  lamda = abs(lamda)-pi
L_2 = L_1+lamda-delta
A_2 = atan(cosu_1*sin(A_1)/(cosu_1*cos(sigma)*cos(A_1)-sinu_1*sin(sigma)))
if sin(A_1)<0 and tan(A_2)>=0:
  A_2 = abs(A_2)
elif sin(A_1)<0 and tan(A_2)<0:
  A_2 = pi-abs(A_2)
elif sin(A_1)>=0 and tan(A_2)>=0:
  A_2 = pi+abs(A_2)
else:
  A_2 = 2*pi-abs(A_2)
B_2 = B_2*180/pi
L_2 = L_2*180/pi
A_2 = A_2*180/pi
print("緯度:",B_2,"度")
print("經度:",L_2,"度")
print("方位角:",A_2,"度")
```

```python
# 大地線反算
a = 6378245
f = 1/298.3
b = a*(1-f)
e2 = 1-b**2/a**2
e_prime2 = (a**2-b**2)/b**2
B_1 = 30
B_1 = B_1*pi/180
L_1 = 107
L_1 = L_1*pi/180
B_2 = 23.438433211916013
B_2 = B_2*pi/180
L_2 = 100.09211190791355
L_2 = L_2*pi/180
W_1 = sqrt(1-e2*(sin(B_1)**2))
W_2 = sqrt(1-e2*(sin(B_2)**2))
sinu_1 = sin(B_1)*sqrt(1-e2)/W_1
cosu_1 = cos(B_1)/W_1
sinu_2 = sin(B_2)*sqrt(1-e2)/W_2
cosu_2 = cos(B_2)/W_2
L_L = L_2-L_1
a_1 = sinu_1*sinu_2
a_2 = cosu_1*cosu_2
b_1 = cosu_1*sinu_2
b_2 = sinu_1*cosu_2
# 首次
p = cosu_2*sin(L_L)
q = b_1-b_2*cos(L_L)
A_1 = atan(p/q)
if p>=0 and q>=0:
  A_1 = abs(A_1)
elif p>=0 and q<0:
  A_1 = pi-abs(A_1)
elif p<0 and q<0:
  A_1 = pi+abs(A_1)
else:
  A_1 = 2*pi-abs(A_1)
sinsigma = p*sin(A_1)+q*cos(A_1)
cossigma = a_1+a_2*cos(L_L)
sigma = atan(sinsigma/cossigma)
if cossigma>=0:
  sigma = abs(sigma)
else:
  sigma = pi-abs(sigma)
sinA_0 = cosu_1*sin(A_1)
cosA_0 = sqrt(1-sinA_0**2)
alpha = (e2/2+e2**2/8+e2**3/16)-(e2**2/16+e2**3/16)*(cosA_0**2)+3*(e2**3)*(cosA_0**4)/128
beta = (e2**2/32+e2**3/32)*(cosA_0**2)-(e2**3/64)*(cosA_0**4)
x = 2*a_1-(cosA_0**2)*cos(sigma)
delta = (alpha*sigma-2*beta*x*sin(sigma))*sinA_0
L = L_L+delta
check = delta*180*3600/pi
print("delta:",check)
former = check+1
# 迭代開始
while abs(check-former)!=0:
  former = check
  p = cosu_2*sin(L)
  q = b_1-b_2*cos(L)
  A_1 = atan(p/q)
  if p>=0 and q>=0:
    A_1 = abs(A_1)
  elif p>=0 and q<0:
    A_1 = pi-abs(A_1)
  elif p<0 and q<0:
    A_1 = pi+abs(A_1)
  else:
    A_1 = 2*pi-abs(A_1)
  sinsigma = p*sin(A_1)+q*cos(A_1)
  cossigma = a_1+a_2*cos(L)
  sigma = atan(sinsigma/cossigma)
  if cossigma>=0:
    sigma = abs(sigma)
  else:
    sigma = pi-abs(sigma)
  sinA_0 = cosu_1*sin(A_1)
  cosA_0 = sqrt(1-sinA_0**2)
  alpha = (e2/2+e2**2/8+e2**3/16)-(e2**2/16+e2**3/16)*(cosA_0**2)+3*(e2**3)*(cosA_0**4)/128
  beta = (e2**2/32+e2**3/32)*(cosA_0**2)-(e2**3/64)*(cosA_0**4)
  x = 2*a_1-(cosA_0**2)*cos(sigma)
  delta = (alpha*sigma-2*beta*x*sin(sigma))*sinA_0
  L = L_L+delta
  check = delta*180*3600/pi
  print("delta:",check)
# 迭代結束
gamma = (cosA_0**4-2*(x**2))*cos(sigma)
k = sqrt(e_prime2*(cosA_0**2))
A = b*(1+k**2/4-3*(k**4)/64+5*(k**6)/256)
B = b*(k**2/8-k**4/32+15*(k**6)/1024)
C = b*(k**4/128-3*(k**6)/512)
S = A*sigma+(2*B*x/(cosA_0**2)+2*C*gamma/(cosA_0**4))*sin(sigma)
A_2 = atan(cosu_1*sin(L)/(b_1*cos(L)-b_2))
A_1 = A_1*180/pi
A_2 = A_2*180/pi
print("方位角:",A_1,"度")
print("方位角:",A_2,"度")
print("大地線:",S,"公尺")
```
