```python
from pykrige import OrdinaryKriging
import matplotlib.pyplot as plt
X_pred = np.linspace(-1, 1, 200)
uk = OrdinaryKriging(X, np.zeros(X.shape), y, variogram_model="gaussian")
y_pred, y_std = uk.execute("grid", X_pred, np.array([0.0]))
y_pred = np.squeeze(y_pred)
y_std = np.squeeze(y_std)
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
ax.scatter(X, y, s=40, label="Input data")
ax.plot(X_pred, y_pred, label="Predicted values")
ax.fill_between(X_pred, y_pred - 3 * y_std, y_pred + 3 * y_std, alpha=0.3, label="Confidence interval",)
ax.legend(loc=9)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(-1, 1)
ax.set_ylim(-2.8, 3.5)
plt.show()
```

```python
from scipy.interpolate import Rbf
import matplotlib.pyplot as plt
func = Rbf(X, y, function="thin-plate")
new_X = np.linspace(-1, 1, 200)
new_y = func(new_X)
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
ax.scatter(X, y, s=40, label="Input data")
ax.plot(new_X, new_y, label="Predicted values")
ax.legend(loc=9)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(-1, 1)
ax.set_ylim(-2.8, 3.5)
plt.show()
```

```python
from scipy import interpolate
import matplotlib.pyplot as plt
func = interpolate.interp1d(X, y, fill_value="extrapolate")
tin_X = np.linspace(-1, 1, 200)
tin_y = func(tin_X)
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
ax.scatter(X, y, s=40, label="Input data")
ax.plot(tin_X, tin_y, label="Predicted values")
ax.legend(loc=9)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(-1, 1)
ax.set_ylim(-2.8, 3.5)
plt.show()
```
